```python
import numpy as np
import matplotlib.pyplot as plt

# ŸÜÿπÿ±ŸÅ ÿßŸÑŸÄ loss function
def loss(w):
    return (w - 3)**2 + 2   # ŸÖŸÜÿ≠ŸÜŸâ ÿ®ÿ≥Ÿäÿ∑ ÿπŸÜÿØŸá minimum ÿπŸÜÿØ W = 3

# ÿßŸÑŸÖÿ¥ÿ™ŸÇÿ© (gradient)
def grad(w):
    return 2 * (w - 3)

# ÿ•ÿπÿØÿßÿØÿßÿ™
w = -5          # ÿßŸÑÿ®ÿØÿßŸäÿ© (ŸÜŸÇÿ∑ÿ© ÿ®ÿπŸäÿØÿ© ÿπŸÜ ÿßŸÑŸÇÿßÿπ)
lr = 0.1        # learning rate
epochs = 20     # ÿπÿØÿØ ÿßŸÑÿÆÿ∑Ÿàÿßÿ™

# ŸÜÿ±ÿ≥ŸÖ ÿßŸÑŸÖŸÜÿ≠ŸÜŸâ
W_values = np.linspace(-5, 10, 200)
L_values = loss(W_values)

plt.figure(figsize=(10,6))
plt.plot(W_values, L_values, label='Loss Function', color='gray')

# Gradient Descent
for i in range(epochs):
    g = grad(w)
    new_w = w - lr * g

    plt.scatter(w, loss(w), color='red')  # ÿßŸÑŸÜŸÇÿ∑ÿ© ÿßŸÑÿ≠ÿßŸÑŸäÿ©
    plt.text(w, loss(w)+1, f'Step {i}', fontsize=8)

    w = new_w

plt.title("Gradient Descent Optimization Path")
plt.xlabel("Weight (W)")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ÿ®ŸäÿßŸÜÿßÿ™ ÿ®ÿ≥Ÿäÿ∑ÿ© (ŸÉÿ£ŸÜŸáÿß dataset)
X = np.array([-4, -3, -2, -1, 0, 1, 2, 3, 4])
Y = (X - 3)**2 + 2   # ŸÜŸÅÿ≥ ŸÅŸÉÿ±ÿ© ÿßŸÑÿØÿßŸÑÿ© ÿßŸÑÿ£ÿµŸÑŸäÿ©

# ÿØÿßŸÑÿ© loss ÿ®ÿ≥Ÿäÿ∑ÿ© (ÿπÿ¥ÿßŸÜ ŸÜÿ≥ÿ™ÿÆÿØŸÖ ŸÉŸÑ ŸÜŸÇÿ∑ÿ© ŸÉÿπŸäŸÜÿ©)
def loss(w, x):
    return (x - 3)**2 + 2 + np.random.randn() * 0.2  # noise ÿ®ÿ≥Ÿäÿ∑

# gradient ÿ™ŸÇÿ±Ÿäÿ®Ÿä (ŸÉÿ£ŸÜŸÜÿß ÿ®ŸÜÿ™ÿπŸÑŸÖ ŸÖŸÜ ÿπŸäŸÜÿ© Ÿàÿßÿ≠ÿØÿ©)
def grad(w, x):
    return 2 * (w - 3)

# ÿ•ÿπÿØÿßÿØÿßÿ™
w = -5
lr = 0.1
epochs = 20

W_values = np.linspace(-5, 10, 200)
L_values = (W_values - 3)**2 + 2

plt.figure(figsize=(10,6))
plt.plot(W_values, L_values, label='Loss Function', color='gray')

for i in range(epochs):
    x_i = np.random.choice(X)  # ŸÜÿÆÿ™ÿßÿ± ÿπŸäŸÜÿ© ÿπÿ¥Ÿàÿßÿ¶Ÿäÿ© ŸÉŸÑ ŸÖÿ±ÿ©
    g = grad(w, x_i)
    w = w - lr * g

    plt.scatter(w, (w - 3)**2 + 2, color='red')
    plt.text(w, (w - 3)**2 + 2 + 0.5, f'Step {i}', fontsize=8)

plt.title("Stochastic Gradient Descent (SGD) Path")
plt.xlabel("Weight (W)")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
